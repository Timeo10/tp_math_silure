# Tp Catfish

# 1. Importer le dataset catfish

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from pmdarima import auto_arima

data = pd.read_csv('catfish.csv')
data.columns = ['date', 'total']

data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

data.head()
data.tail()
```

# 2. Tracer la courbe de vente de poisson-chats

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['total'], label='Ventes de Poisson-Chats')
plt.title('Ventes de Poisson-Chats de 1986 √† 2012')
plt.xlabel('Ann√©e')
plt.ylabel('Ventes (en milliers de livres)')
plt.legend()
plt.show()
```

# 3. Tracer la courbe de vente de 1996 √† 2000

```python
data_96_00 = data['1996-01-01':'2000-12-31']
plt.figure(figsize=(12, 6))
plt.plot(data_96_00.index, data_96_00['total'], label='Ventes de Poisson-Chats (1996-2000)')
plt.title('Ventes de Poisson-Chats de 1996 √† 2000')
plt.xlabel('Ann√©e')
plt.ylabel('Ventes (en milliers de livres)')
plt.legend()
plt.show()

```

# 4. Quelle est la moyenne des ventes


```python
mean_92 = data['1992-01-01':'1992-12-31']['total'].mean()

mean_90s = data['1990-01-01':'1999-12-31']['total'].mean()

mean_total = data['total'].mean()

print('Ann√©e 1992:',mean_92)
print('D√©c√©nie 1990 - 99:',mean_90s)
print('Total:',mean_total)
```

# 5. Quelle est la m√©diane des ventes - sur l‚Äôann√©e 92 ? - sur la d√©cennie 90 (1990 √† 1999) ? - sur l‚Äôensemble du dataset ?

```python
median_92 = data['1992-01-01':'1992-12-31']['total'].median()

median_90s = data['1990-01-01':'1999-12-31']['total'].median()

median_total = data['total'].median()

print('Ann√©e 1992:',median_92)
print('D√©c√©nie 1990 - 99:',median_90s)
print('Total:',median_total)
```

# 6. Quelle est le mode des ventes

```python
mode_92 = data['1992-01-01':'1992-12-31']['total'].mode()

mode_90s = data['1990-01-01':'1999-12-31']['total'].mode()

mode_total = data['total'].mode()

mode_92, mode_90s, mode_total
```

# 7. Quelle est l‚Äô√©cart-type des ventes

```python

std_92 = data['1992-01-01':'1992-12-31']['total'].std()

std_90s = data['1990-01-01':'1999-12-31']['total'].std()

std_total = data['total'].std()

print('Ann√©e 1992:',std_92)
print('D√©c√©nie 1990 - 99:',std_90s)
print('Total:',std_total)
```

# 8. La s√©rie est-elle stationnaire ?

```python
result = adfuller(data['total'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
for key, value in result[4].items():
    print('Critique valeur ({}): {}'.format(key, value))
```

# 9. La s√©rie comporte elle une saisonnalit√© ?

```python
decomposition = sm.tsa.seasonal_decompose(data['total'], model='additive')
fig = decomposition.plot()
plt.show()
```

# 10. La s√©rie pr√©sente elle une tendance ?

```python
trend = decomposition.trend
plt.figure(figsize=(12, 6))
plt.plot(data.index, trend, label='Tendance')
plt.title('Tendance des Ventes de Poisson-Chats')
plt.xlabel('Ann√©e')
plt.ylabel('Tendance des Ventes')
plt.legend()
plt.show()
```

# 11. Pouvez vous √† partir des donn√©es de 1986 √† 2000 √©tablir un mod√®le pr√©dictif sur les ventes de 2001 √† 2012 ? Comparez le avec les ventes r√©elles et critiquez le

```python
train = data[:'2000-12-31']
test = data['2001-01-01':]

model = auto_arima(train['total'], seasonal=True, m=12)
model.fit(train['total'])

forecast = model.predict(n_periods=len(test))

plt.figure(figsize=(12, 6))
plt.plot(train.index, train['total'], label='Train')
plt.plot(test.index, test['total'], label='Test')
plt.plot(test.index, forecast, label='Pr√©dictions')
plt.title('Pr√©diction des Ventes de Poisson-Chats de 2001 √† 2012')
plt.xlabel('Ann√©e')
plt.ylabel('Ventes (en milliers de livres)')
plt.legend()
plt.show()
```

# 12. Quel mod√®le nous propose la libraire pmdarima ?

```python
model_summary = model.summary()
print(model_summary)
```

# 5.2 Poisson-chat et loi normale

# 1. Entre 10 et 25 cm ?

```python
from statistics import NormalDist

mu = 25
sigma = 7.5

dist = NormalDist(mu, sigma)
prob_10_25 = dist.cdf(25) - dist.cdf(10)
print(f"Probabilit√© entre 10 et 25 cm : {prob_10_25:.3f}")
```

# 2. Moins de 25 cm ?

```python
mu = 25
sigma = 7.5

dist = NormalDist(mu, sigma)

prob_less_25 = dist.cdf(25)
print(f"Probabilit√© moins de 25 cm : {prob_less_25:.3f}")
```

# 3. Plus de 30 cm ?

```python
mu = 25
sigma = 7.5

# Cr√©ation de la distribution normale
dist = NormalDist(mu, sigma)

# Calcul de la probabilit√© de plus de 30 cm
prob_more_30 = 1 - dist.cdf(30)

# Affichage du r√©sultat
print(f"Probabilit√© plus de 30 cm : {prob_more_30:.3f}")
```
